# Dataset Source

Provided: 'mymoviedb.csv' (~9,837 movies from TMDB).
Columns: Release_Date, Title, Overview, Popularity, Vote_Count, Vote_Average, Original_Language, Genre (comma-separated), Poster_Url.
Notes: All entries are movies; no TV shows or runtimes. Release dates range from historical to recent (e.g., 2022+).

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np  # For any numerical operations

# Load the dataset with fixes for ParserError (use Python engine, skip bad lines)
df = pd.read_csv('mymoviedb.csv', engine='python', on_bad_lines='skip')

# Display head and info to verify
print(df.head())
df.info()
```

# Data Overview and Cleaning

```python
# Convert Release_Date to datetime and extract year
df['Release_Date'] = pd.to_datetime(df['Release_Date'], errors='coerce')
df['release_year'] = df['Release_Date'].dt.year

# Handle any missing values (e.g., fill NaNs in strings with 'Unknown')
df.fillna({'Title': 'Unknown', 'Overview': 'Unknown', 'Genre': 'Unknown', 'Original_Language': 'Unknown'}, inplace=True)

# Drop duplicates if any (based on Title)
df.drop_duplicates(subset='Title', keep='first', inplace=True)

# Display cleaned info
df.info()
print(df[['release_year', 'Vote_Average', 'Popularity']].describe())
```

# Explore Counts by Type (Movie/TV), Year Trends, and Top Genres

```python
# Counts by type (adapted: all movies)
print("Counts by Type: All are Movies -", len(df), "(No TV shows in dataset)")

# Year trends (releases per year)
year_trends = df.groupby('release_year')['Title'].count().reset_index(name='count')
year_trends = year_trends[year_trends['release_year'] > 2000]  # Focus on recent
print("\nYear Trends (Releases):\n", year_trends.tail(10))

# Top genres (split and count)
df_genres = df.copy()
df_genres['Genre'] = df_genres['Genre'].str.split(', ')
df_genres = df_genres.explode('Genre')
genre_counts = df_genres['Genre'].value_counts().head(20)
print("\nTop Genres:\n", genre_counts)
```

# Visualize Content Growth Over Time and Runtime Distributions

```python
# Content growth over time (by release year)
plt.figure(figsize=(12, 6))
sns.lineplot(data=year_trends, x='release_year', y='count', marker='o')
plt.title('Movie Release Growth Over Time')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
plt.grid(True)
plt.show()

# Adapted distributions (Vote_Average and Popularity, since no runtime)
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='Vote_Average', bins=20, kde=True, color='blue', label='Vote Average')
plt.title('Vote Average Distribution')
plt.xlabel('Vote Average')
plt.ylabel('Frequency')
plt.legend()
plt.show()

plt.figure(figsize=(12, 6))
sns.histplot(data=df[df['Popularity'] < 100], x='Popularity', bins=30, kde=True, color='green', label='Popularity (capped at 100 for visibility)')
plt.title('Popularity Distribution (Capped)')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

# Generate Top-10 Lists (Most Common Genres, Years)

```python
# Top-10 Genres
top10_genres = genre_counts.head(10).reset_index(name='count')
top10_genres.columns = ['Genre', 'Count']
print("Top-10 Genres:\n", top10_genres.to_markdown(index=False))

# Top-10 Years (most releases)
top10_years = year_trends.sort_values('count', ascending=False).head(10)
print("\nTop-10 Years by Releases:\n", top10_years.to_markdown(index=False))
```

# Export Visual Report (Plots + Summary)

```python
# Save plots
plt.figure(figsize=(12, 6))
sns.lineplot(data=year_trends, x='release_year', y='count', marker='o')
plt.title('Movie Release Growth Over Time')
plt.savefig('movie_growth.png')

plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='Vote_Average', bins=20, kde=True, color='blue')
plt.title('Vote Average Distribution')
plt.savefig('vote_average_dist.png')

# Summary report
summary = f"""
# TMDB Movie EDA Summary (Adapted)
- Total Movies: {len(df)} (No TV Shows)
- Top Genre: {top10_genres['Genre'].iloc[0]} ({top10_genres['Count'].iloc[0]} movies)
- Peak Year: {top10_years['release_year'].iloc[0]} ({top10_years['count'].iloc[0]} releases)
- Average Vote: {df['Vote_Average'].mean():.1f}
- Median Popularity: {df['Popularity'].median():.1f}
- Insights: Dataset shows growth in releases post-2010; focus on high-vote genres like Drama for acquisitions.
"""
with open('tmdb_movie_eda_summary.md', 'w') as f:
    f.write(summary)

print("Report exported: movie_growth.png, vote_average_dist.png, tmdb_movie_eda_summary.md")
```
