# Dataset Source

From Kaggle: https://www.kaggle.com/datasets/shivamb/netflix-shows (updated monthly, includes data up to recent years).
Columns: show_id, type (Movie/TV Show), title, director, cast, country, date_added, release_year, rating, duration, listed_in (genres), description.
~8,800 entries; mix of movies (~70%) and TV shows (~30%).

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np  # For any numerical operations

# Load the dataset (download from Kaggle or use raw GitHub URL)
#url = 'https://raw.githubusercontent.com/shivamb/netflix-shows/master/netflix_titles.csv'
df = pd.read_csv('mymoviedb.csv')  # Or pd.read_csv('netflix_titles.csv') if local

# Display head and info to verify
print(df.head())
df.info()
```

# Data Overview and Cleaning

```python
# Handle missing values
df.fillna({'director': 'Unknown', 'cast': 'Unknown', 'country': 'Unknown', 'date_added': 'Unknown', 'rating': 'Unknown', 'duration': 'Unknown'}, inplace=True)

# Convert date_added to datetime and extract year
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['added_year'] = df['date_added'].dt.year

# Parse duration to numeric
df['duration_num'] = df['duration'].str.extract('(\d+)').astype(float)  # Extract numbers
df['duration_unit'] = df['duration'].str.extract('([a-zA-Z]+)')  # Units: min/Season/Seasons

# Drop duplicates if any
df.drop_duplicates(subset='title', keep='first', inplace=True)

# Display cleaned info
df.info()
print(df[['type', 'release_year', 'added_year', 'duration_num', 'duration_unit']].head())
```
